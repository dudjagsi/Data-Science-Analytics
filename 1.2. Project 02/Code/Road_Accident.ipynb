# Graph of total accidents and how many people killed n injured

```python
plt.figure(figsize=(10,7))
sns.lineplot(df.iloc[:,0],df.iloc[:,1],color="red")
sns.lineplot(df.iloc[:,0],df.iloc[:,2],color="green")
sns.lineplot(df.iloc[:,0],df.iloc[:,3],color="blue")
plt.legend(["Accidents","killed","injured"])
plt.xlabel('years')
plt.ylabel('Persons')
plt.grid()
plt.show()
```

```python
graph=df[["Years","Total Number of Road Accidents (in numbers)","Total Number of Persons Killed (in numbers)","Total Number of Persons Injured (in numbers)"]]
```

```python
graph.head()
```

```python
graph.columns=["years","accidents","killed","injured"]
```

```python
graph.head()
```

```python

x=np.array([1970, 1980, 1990, 1994, 1995, 1996, 1997, 1998, 1999, 2000, 2001,
        2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012,
        2013, 2014, 2015, 2016]
```

```python
accident=graph['accidents']
killed=graph['killed']
injured=graph['injured']
```

```python
plt.figure(figsize = (20,10))
plt.rcParams.update({'font.size':18})
plt.bar(x-0.2,accident,label='accidents',width=0.2,align='center')
plt.bar(x,killed,label='killed',width=0.2,align='center')
plt.bar(x+0.2,injured,label='injured',width=0.2,align='center')
plt.legend(['accidents','killed','injured'])
plt.xticks(x, rotation = 'vertical' )
plt.show()
```
